```python
# ============================================
# FASE 1.12: AN√ÅLISIS PARA VERSI√ìN REDUCIDA
# ============================================

# Calcular tama√±o √≥ptimo para la versi√≥n reducida
# Recomendaci√≥n: entre 10-30% del dataset original, dependiendo del tama√±o

dataset_size = len(df)
print("=" * 80)
print("AN√ÅLISIS PARA VERSI√ìN REDUCIDA DEL DATASET")
print("=" * 80)

print(f"\nüìä Tama√±o del dataset original: {dataset_size:,} observaciones")

# Sugerencias de tama√±os
suggested_sizes = {
    '10%': int(dataset_size * 0.10),
    '15%': int(dataset_size * 0.15),
    '20%': int(dataset_size * 0.20),
    '25%': int(dataset_size * 0.25),
    '30%': int(dataset_size * 0.30)
}

print("\nüí° Tama√±os sugeridos para la versi√≥n reducida:")
for percentage, size in suggested_sizes.items():
    print(f"   - {percentage:5s}: {size:,} observaciones")

# Recomendaci√≥n basada en el tama√±o del dataset
if dataset_size < 5000:
    recommended_percentage = 0.30
elif dataset_size < 50000:
    recommended_percentage = 0.20
else:
    recommended_percentage = 0.15

recommended_size = int(dataset_size * recommended_percentage)
print(f"\n‚úÖ Tama√±o recomendado: {recommended_percentage*100:.0f}% = {recommended_size:,} observaciones")
print(f"   (Mantiene representatividad y es manejable para an√°lisis detallado)")
```

```python
# ============================================
# FASE 1.13: FUNCI√ìN PARA CREAR VERSI√ìN REDUCIDA ESTRATIFICADA
# ============================================

# Esta funci√≥n crea un subset estratificado del dataset
def create_stratified_sample(data, target_col, sample_size=None, sample_fraction=0.20, random_state=42):
    """
    Crea una muestra estratificada del dataset manteniendo la distribuci√≥n de la variable objetivo
    
    Par√°metros:
    -----------
    data : DataFrame
        Dataset original
    target_col : str
        Nombre de la columna objetivo
    sample_size : int, opcional
        N√∫mero espec√≠fico de observaciones para la muestra
    sample_fraction : float, opcional
        Fracci√≥n del dataset original (usar si sample_size es None)
    random_state : int
        Semilla para reproducibilidad
    
    Retorna:
    --------
    DataFrame : Muestra estratificada del dataset
    """
    from sklearn.model_selection import train_test_split
    
    # Calcular tama√±o de muestra
    if sample_size is None:
        sample_size = int(len(data) * sample_fraction)
    
    # Calcular la fracci√≥n para train_test_split
    test_fraction = sample_size / len(data)
    
    # Crear muestra estratificada
    _, sample = train_test_split(
        data,
        test_size=test_fraction,
        stratify=data[target_col],
        random_state=random_state
    )
    
    return sample

print("‚úì Funci√≥n create_stratified_sample definida correctamente")
```

```python
# ============================================
# FASE 1.14: GENERAR VERSI√ìN REDUCIDA DEL DATASET
# ============================================

# Importar librer√≠a necesaria
from sklearn.model_selection import train_test_split

# Par√°metros para la versi√≥n reducida
SAMPLE_FRACTION = 0.20  # Modificar seg√∫n necesidad (0.10 = 10%, 0.20 = 20%, etc.)
RANDOM_STATE = 42       # Para reproducibilidad

# Generar la versi√≥n reducida estratificada
df_reduced = create_stratified_sample(
    data=df,
    target_col=target_column,
    sample_fraction=SAMPLE_FRACTION,
    random_state=RANDOM_STATE
)

print("=" * 80)
print("VERSI√ìN REDUCIDA GENERADA EXITOSAMENTE")
print("=" * 80)

print(f"\nüìä Dataset original: {len(df):,} observaciones")
print(f"üìä Dataset reducido: {len(df_reduced):,} observaciones")
print(f"üìä Porcentaje conservado: {len(df_reduced)/len(df)*100:.2f}%")
```

```python
# ============================================
# FASE 1.15: VERIFICACI√ìN DE LA ESTRATIFICACI√ìN
# ============================================

# Comparar distribuci√≥n de la variable objetivo entre dataset original y reducido
print("\nüîç VERIFICACI√ìN DE LA ESTRATIFICACI√ìN:")
print("=" * 80)

# Crear DataFrame comparativo
comparison = pd.DataFrame({
    'Original_Count': df[target_column].value_counts().sort_index(),
    'Original_%': (df[target_column].value_counts(normalize=True) * 100).sort_index(),
    'Reduced_Count': df_reduced[target_column].value_counts().sort_index(),
    'Reduced_%': (df_reduced[target_column].value_counts(normalize=True) * 100).sort_index()
})

comparison['Difference_%'] = (comparison['Reduced_%'] - comparison['Original_%']).abs()

print("\nüìä Comparaci√≥n de distribuciones:")
display(comparison.round(2))

print(f"\n‚úÖ Diferencia promedio: {comparison['Difference_%'].mean():.4f}%")
print(f"‚úÖ Diferencia m√°xima: {comparison['Difference_%'].max():.4f}%")

if comparison['Difference_%'].max() < 1.0:
    print("\n‚úÖ EXCELENTE: La estratificaci√≥n mantiene muy bien la distribuci√≥n original (<1% diferencia)")
elif comparison['Difference_%'].max() < 2.0:
    print("\n‚úÖ BUENO: La estratificaci√≥n mantiene bien la distribuci√≥n original (<2% diferencia)")
else:
    print("\n‚ö†Ô∏è ACEPTABLE: Hay algunas diferencias en la distribuci√≥n, pero son tolerables")
```

```python
# ============================================
# FASE 1.16: VISUALIZACI√ìN COMPARATIVA
# ============================================

# Crear visualizaci√≥n comparativa de las distribuciones
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Dataset original
df[target_column].value_counts().sort_index().plot(
    kind='bar', 
    ax=axes[0], 
    color='steelblue', 
    edgecolor='black',
    alpha=0.8
)
axes[0].set_title(f'Distribuci√≥n Original\n(n={len(df):,})', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Clases', fontsize=12)
axes[0].set_ylabel('Frecuencia', fontsize=12)
axes[0].grid(axis='y', alpha=0.3)

# A√±adir etiquetas con valores
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%d', padding=3)

# Dataset reducido
df_reduced[target_column].value_counts().sort_index().plot(
    kind='bar', 
    ax=axes[1], 
    color='coral', 
    edgecolor='black',
    alpha=0.8
)
axes[1].set_title(f'Distribuci√≥n Reducida\n(n={len(df_reduced):,})', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Clases', fontsize=12)
axes[1].set_ylabel('Frecuencia', fontsize=12)
axes[1].grid(axis='y', alpha=0.3)

# A√±adir etiquetas con valores
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%d', padding=3)

plt.tight_layout()
plt.show()

# Interpretaci√≥n: Las barras deben mantener proporciones similares entre ambos gr√°ficos
```
